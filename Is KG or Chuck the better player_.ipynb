# Charles Barkley recently teased Kevin Garnett for being a borderline hall of famer. Kevin Garnett will be inducted into the Hall of Fame May 13-15 , 2021. We will look at the data to see who is the better player between two of the best Power Forwards of all-time, Kevin Garnett and Charles Barkely.

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sklearn
sns.set()
```

```python
# Import Data
kg_df = pd.read_csv('KGData.csv')
cb_df = pd.read_csv('BarkleyData.csv')
```

```python
# See what type of statistics we can use
kg_df.columns
```

```python
# Create the Free Throw Rate Statistic
kg_df['FTM'] = (kg_df['FT%']*kg_df['FTA'])
kg_df['FTR'] = kg_df['FTM']/kg_df['FGA']

cb_df['FTM'] = (cb_df['FT%']*cb_df['FTA'])
cb_df['FTR'] = cb_df['FTM']/cb_df['FGA']
```

```python
# Take the last row to get Career Averages for both players.
```

```python
kg_avgs = kg_df.tail(1)
cb_avgs = cb_df.tail(1)
kg_avgs = kg_avgs.drop(['Season','Age','Tm','Lg','Pos'],axis = 1)
cb_avgs = cb_avgs.drop(['Season','Age','Tm','Lg','Pos'],axis = 1)
```

# Let's answer the question of who was the better scorer
    Offensive stats we will compare will be Points, Effective Field Goal Percentage, and Free Throw Rate

```python
# Offensive Statistics
kg_ppg = kg_df['PTS'][:-1]
kg_efg = kg_df['eFG%'][:-1]
kg_ftr = kg_df['FTR'][:-1]

# Age
kg_age = kg_df['Age'][:-1].astype(int)
cb_age = cb_df['Age'][:-1].astype(int)

cb_ppg = cb_df['PTS'][:-1]
cb_ftr = cb_df['FTR'][:-1]
cb_efg = cb_df['eFG%'][:-1]

# Defensive Statistics
kg_bpg = kg_df['BLK'][:-1]
kg_spg = kg_df['STL'][:-1]

cb_bpg = cb_df['BLK'][:-1]
cb_spg = cb_df['STL'][:-1]
```

```python
# Offensive Comparison Plot
import matplotlib

font = {'fontname': 'Georgia'}

fig, (ax1, ax2,ax3) = plt.subplots(3, figsize = (15,10))



# Points Per Game Plot
ax1.set_title('Was Kevin Garnett or Charles Barkley a better scorer?',**font,fontsize = 25)

ax1.plot(kg_age,kg_ppg, label = 'Kevin Garnett',color = 'Orange')
ax1.plot(cb_age,cb_ppg, label = 'Charles Barkley',color = 'Navy')


fig.text(0.001, 0.8, 'Points Per Game', va = 'center', rotation = 'vertical',fontsize = 20,**font)


# Effective Field Goal Percentage

ax2.plot(kg_age,kg_efg, label = 'Kevin Garnett',color = 'Orange')
ax2.plot(cb_age,cb_efg, label = 'Charles Barkley',color = 'Navy')
ax2.legend(loc = 'upper right',prop={"size":20},bbox_to_anchor=(1.35,1))

fig.text(0.001, 0.5, 'Effective FG%', va = 'center', rotation = 'vertical',fontsize = 20,**font)

# Free Throw Rate = FTM/FGA


ax3.plot(kg_age,kg_ftr, label = 'Kevin Garnett',color = 'Orange')
ax3.plot(cb_age,cb_ftr, label = 'Charles Barkley',color = 'Navy')


fig.text(0.001, 0.2, 'Free Throw Rate', va = 'center', rotation = 'vertical',fontsize = 20,**font)


fig.text(0.4,0.001,'Age',ha = 'center',fontsize = 25,**font)
fig.tight_layout(pad=3.5)

ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis = 'y', labelsize = 15)
ax2.tick_params(axis='x', labelsize=15)
ax2.tick_params(axis = 'y', labelsize = 15)
ax3.tick_params(axis='x', labelsize=15)
ax3.tick_params(axis = 'y', labelsize = 15)
```

# Stats Description

    1. Points Per Game is how many points scored a player scores per game
    2. Effective Field Goal Percentage captures the percentage of shots a player makes. It gives 3 pointers 50% more credit than a 2 pointer because it's worth 50% more.
    3. Free throw rate tracks how often a player goes to the foul line as well as their free throw percentage
     

# Who is the better scorer?
    
From the plot here, **Charles Barkley is clearly the better scorer**. His peak points per game were higher than any of Kevin Garnetts best seasons, His Effective Field Goal Percentage was higher for most of their careers. His free throw rate was also higher for their entire careers.

```python
# Defensive Comparison Plot

font = {'fontname': 'Georgia'}

fig, (ax1, ax2) = plt.subplots(2, figsize = (15,10))



# Blocks Per Game Plot
ax1.set_title('Was Kevin Garnett or Charles Barkley a better defender?',**font,fontsize = 25)

ax1.plot(kg_age,kg_bpg, label = 'Kevin Garnett',color = 'Orange')
ax1.plot(cb_age,cb_bpg, label = 'Charles Barkley',color = 'Navy')
ax1.legend(loc = 'upper right',prop={"size":20})

fig.text(0.001, 0.7, 'Blocks Per Game', va = 'center', rotation = 'vertical',fontsize = 20,**font)


# Steals Per Game

ax2.plot(kg_age,kg_spg, label = 'Kevin Garnett',color = 'Orange')
ax2.plot(cb_age,cb_spg, label = 'Charles Barkley',color = 'Navy')
ax2.legend(loc = 'upper right',prop={"size":20})

fig.text(0.001, 0.3, 'Steals Per Game', va = 'center', rotation = 'vertical',fontsize = 20,**font)







fig.text(0.5,0.001,'Age',ha = 'center',fontsize = 25,**font)
fig.tight_layout(pad=3.5)

ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis = 'y', labelsize = 15)
ax2.tick_params(axis='x', labelsize=15)
ax2.tick_params(axis = 'y', labelsize = 15)
```

Charles Barkley averaged more steals than Kevin Garnett at the peak of his career but Kevin Garnett had a few seasons averaging more steals per game than Charles Barkley. We have to use blocks and steals per game because we don't have stats such as oppenents field goal percentage because they weren't tracked until recently. Based of steals and blocks, we would have to conclude that **Kevin Garnett was a better defender overall** because he average signficantly more blocks per game than Charles for his entire career and was atleast comparable to Charles in steals per game.
